**Q1.** Write a Python program that defines a function to check whether a given year is a leap year. The program should take input from the user, call the function, and print whether the entered year is a leap year or not.(ZERO SHOT EXAMPLE)

```python
def is_leap_year(year):
    """
    Checks if a given year is a leap year.

    A year is a leap year if:
    - It is divisible by 4, but not by 100
    - OR it is divisible by 400
    """
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Take input from the user
try:
    year_input = int(input("Enter a year: "))

    # Call the function and print the result
    if is_leap_year(year_input):
        print(f"{year_input} is a leap year.")
    else:
        print(f"{year_input} is not a leap year.")
except ValueError:
    print("Invalid input. Please enter a valid year (a whole number).")
```

**Q2**.write a Python program that defines a function to convert centimeters to inches.
The program should take input from the user in centimeters, call the function, and print the converted value in inches.

Example:
Input: 10

Output: 3.937 inches

```python
def cm_to_inches(cm):
    """
    Converts a value from centimeters to inches.

    1 inch = 2.54 centimeters
    """
    inches = cm * 0.393701 # Approximately 1 inch = 2.54 cm, so 1 cm = 0.393701 inches
    return inches

# Take input from the user
try:
    cm_input = float(input("Enter a value in centimeters: "))

    # Call the function and print the result
    converted_inches = cm_to_inches(cm_input)
    print(f"{cm_input} centimeters is equal to {converted_inches:.3f} inches.")
except ValueError:
    print("Invalid input. Please enter a valid number for centimeters.")
```

**Q3.** Write a Python program that defines a function to format a full name as “Last, First”.
The program should take a full name as input from the user, call the function, and print the formatted name.

Examples:
Input: "Rimsha Mujeeb"
Output: "Mujeeb, Rimsha"

Input: "Priya Raj"
Output: "Raj, Priya"

Input: "Pujitha sham"
Output: "sham, Pujitha"

```python
def format_name(full_name):
    """
    Formats a full name as 'Last, First'.
    Assumes the full name is given as 'First Last'.
    """
    parts = full_name.split()
    if len(parts) >= 2:
        first_name = parts[0]
        last_name = parts[-1]
        return f"{last_name}, {first_name}"
    else:
        return full_name # Return as is if not enough parts to split

# Take input from the user
name_input = input("Enter a full name (e.g., John Doe): ")

# Call the function and print the result
formatted_name = format_name(name_input)
print(f"Formatted Name: \"{formatted_name}\"")
```

# ZERO SHOT PROMPT

Write a Python function that counts the number of vowels (a, e, i, o, u) in a given string.
The program should take input from the user and print the number of vowels.

```python
def count_vowels(input_string):
    """
    Counts the number of vowels (a, e, i, o, u) in a given string.
    The count is case-insensitive.
    """
    vowels = "aeiou"
    vowel_count = 0
    for char in input_string:
        if char.lower() in vowels:
            vowel_count += 1
    return vowel_count

# Take input from the user
user_string = input("Enter a string: ")

# Call the function and print the result
num_vowels = count_vowels(user_string)
print(f"The number of vowels in the string is: {num_vowels}")
```

**Q5.**Write a Python program that defines a function to read a .txt file and return the number of lines it contains.
The program should take the file name as input from the user, call the function, and print the total number of lines.

```python
def count_lines_in_file(file_path):
    """
    Reads a text file and returns the number of lines it contains.
    Handles FileNotFoundError if the file does not exist.
    """
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()
            return len(lines)
    except FileNotFoundError:
        return -1  # Indicate that the file was not found
    except Exception as e:
        print(f"An error occurred: {e}")
        return -2 # Indicate other errors

# Take file name input from the user
file_name = input("Enter the name of the .txt file (e.g., my_document.txt): ")

# Call the function and print the result
line_count = count_lines_in_file(file_name)

if line_count == -1:
    print(f"Error: The file '{file_name}' was not found.")
elif line_count == -2:
    print(f"An unexpected error occurred while reading '{file_name}'.")
elif line_count is not None:
    print(f"The file '{file_name}' contains {line_count} lines.")
```
